## Dictionaries

They are unordered collections of items. They store data in key value pairs. Keys must be unique and immutable

```python
## Creating dictionaries
emptyDict = {}
print(type(emptyDict))
```

```python
emptyDict = dict()
print(type(emptyDict))
```

```python
student = {'name' : 'helm', 'age' : 45, 'grade' : 76}
print(student)
```

```python
## Accessing dictionary elements
student = {'name': 'jack', 'age': 22, 'grade': 98}
print(student['age'])
print(student['grade'])

print(student.get('grade'))
print(student.get('lastName'))
```

```python
student['age'] = 20
print(student)
student['address'] = 'India' # added a new key
print(student)
```

```python
del student['grade']
print(student)
```

```python
## Dictionary methods

keys = student.keys()
print(keys)
print(student.values())

items = student.items() ## get all key value pairs
print(items)
```

```python
## shallow copy

studentCopy = student
print(student)
print(studentCopy)
```

```python
student['name'] = 'Ram'
print(student)
print(studentCopy)
```

```python
studentCopy1 = student.copy() ## shallow copy -  allocates a different memory for newly created variable
print(studentCopy1)
student['name'] = 'Lucky'
print(student)
print(studentCopy1)
```

```python
## Iterating over dictionaries
for keys in student.keys():
    print(keys)
```

```python
for key, value in student.items():
    print(f'{key} : {value}')
```

```python
##Nested dictionary

students = {
    'student1': {'name': 'Roy', 'age': 21},
    'student2': {'name': 'Henry', 'age': 24}
}
print(students)
```

```python
## Access nested dictionary elements
print(students['student2']['name'])
```

```python
for studentId, studentInfo in students.items():
    print(f'{studentId} : {studentInfo}')
    for key, value in studentInfo.items():
        print(f'{key}: {value}')
```

```python
## Dictionary Comprehension

squares = {x: x ** 2 for x in range(5)}
print(squares)
```

```python
## Conditional dictionary comprehension

evens = {x: x ** 2 for x in range(10) if x % 2 == 0}
print(evens)
```

```python
## Practical Examples
# Use a dictionary to frequency of elements in list
numbers = [1, 2, 2, 3, 3, 4, 4, 4, 4, 4]
frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1
print(frequency)
```

```python
## Merge 2 dictionaries in open
d1 = {'a' : 1, 'b' : 2}
d2 = {'b': 3, 'c': 4}
mDict = {**d1, **d2}
print(mDict)
```

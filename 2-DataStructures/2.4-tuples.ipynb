## Tuples
Tuples are ordered collections of items that are immutable. They are similar to list but their immutability makes them different 

```python
## Creating a tuple

emptyTuple = ()
print(type(emptyTuple))
```

```python
tpl = tuple()
print(type(tpl))
```

```python
numbers = tuple([1, 2, 3, 4, 5, 6])
print(numbers)
```

```python
print(numbers[0])
```

```python
## Immutable nature of tuples

lst = [1, 2, 3, 4, 5]
print(lst)
lst[1] = 9
print(lst)
```

```python
numbers[1] = 89
```

```python
packedTuple = 1, 5, 8, 9
print(packedTuple)
```

```python
## iterating over nested tuple
nestedTuple = ((1, 2, 3), (4, 5, 6), (True, False))
for subTuple in nestedTuple:
    for item in subTuple:
        print(item, end=' ')
```

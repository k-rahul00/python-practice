### Sets
Sets are built in data type in Python used to store collection of unique items. They are unordered and doesn't allow any duplicates.
They are useful in eliminating duplicates, performing mathematical set operations like union, intersection, difference and symmetric difference

```python
## Creating a set
mySet = {1, 2, 3, 4, 5}
print(mySet)
print(type(mySet))
```

```python
myEmptySet = set()
print(type(myEmptySet))
```

```python
myEmptySet = set([1, 2, 3, 4, 5, 0, 5, 5, 4, 6, 3, 6])
print(myEmptySet)
print(type(myEmptySet))
```

```python
## Basic set operations

# Adding and removing elements
mySet.add(7)
print(mySet)
mySet.remove(3)
print(mySet)
```

```python
mySet.remove(10)
```

```python
mySet.discard(11)
print(mySet)
```

```python
#pop method

removedElement = mySet.pop()
print(removedElement)
print(mySet)
```

```python
## clear all the elements
mySet.clear()
print(mySet)
```

```python
## set membership tests
mySet = {1, 2, 3, 4, 5}
print(3 in mySet)
print(10 in mySet)
```

```python
## Mathematical Operations
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

## Union
unionSet = set1.union(set2)
print(unionSet)

## Intersection
intersectionSet = set1.intersection(set2)
print(intersectionSet)

set1.intersection_update(set2)
print(set1)
```

```python
## difference
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

print(set1.difference(set2))
```

```python
set1.symmetric_difference(set2)
```

```python
## sets methods

set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5}

# is subset
print(set1.issubset(set2))
print(set1.issuperset(set2))
```

```python
lst = [ 1, 2, 3, 4, 5, 4, 5, 6, 7]
set(lst)
```

```python
## counting unique words in text

text = 'in this tutorial we are discussing about python'
words = text.split()

uniqueWords = set(words)
print(uniqueWords)
print(len(uniqueWords))
```

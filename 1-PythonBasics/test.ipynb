Python Basics Semantics and Syntax ( Syntax is aabout the correct arrangements of symbols in a code. Semantics refers to the interpretation of the symabols chars and commands in a language - what the code is supposed to do when it runs)

Basic Syntax rules in Python
1. Python is case sensitive

```python
name = 'Kelly'
Name = 'Kremlin'

print(name)
print(Name)
```

2. Indentation - Python uses indentation to define the structure and hierarchy of the code. Unlike many other programming languages that uses braces{}.

```python
age = 32
if age < 30:
    print('user is in twenties')
else:
    print('user is in 30s')
```

3. Line Continuation - 

```python
total = 1 + 2 + 3 + 4 + 5 + \
6 + 7
print(total)
```

4. Multiple Statements on a single line

```python
x = 10 ; y = 18 ; ans = x + y
print(ans)
```
